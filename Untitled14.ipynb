```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Cargar el archivo
# Asegúrate de que la ruta sea correcta para tu carpeta
archivo = 'archive/RickAndMortyScripts.csv'
df = pd.read_csv(archivo)

# 2. Preprocesamiento
# Creamos la columna 'word_count' necesaria para los gráficos A, B y D
df['word_count'] = df['line'].apply(lambda x: len(str(x).split()))

# Configuración de estilo general
sns.set_style("whitegrid")

# ========================================================
# A) HISTOGRAMA (Distribución)
# ========================================================
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='word_count', kde=True, color='teal', bins=30)
plt.title('A) Histograma: Distribución de la Longitud de Diálogos')
plt.xlabel('Número de Palabras')
plt.ylabel('Frecuencia')
plt.show()

# ========================================================
# B) BOXPLOT (Outliers y Mediana)
# ========================================================
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['word_count'], color='salmon')
plt.title('B) Boxplot: Detección de Outliers')
plt.xlabel('Cantidad de Palabras')
plt.show()

# ========================================================
# C) GRÁFICO DE BARRAS (Categoría - CORREGIDO)
# ========================================================
plt.figure(figsize=(12, 6))

# Filtramos los top 10 personajes
top_personajes = df['name'].value_counts().nlargest(10)

# Solución del warning: asignamos 'x' también a 'hue' y desactivamos la leyenda
sns.barplot(
    x=top_personajes.index, 
    y=top_personajes.values, 
    hue=top_personajes.index, 
    legend=False, 
    palette='viridis'
)

plt.title('C) Gráfico de Barras: Top 10 Personajes con más Líneas')
plt.ylabel('Cantidad de Líneas')
plt.xticks(rotation=45) # Rotamos los nombres para que se lean bien
plt.show()

# ========================================================
# D) GRÁFICO DE DISPERSIÓN (Tendencia)
# ========================================================
plt.figure(figsize=(12, 6))
# Usamos el índice del dataframe como eje X (tiempo/secuencia)
sns.scatterplot(x=df.index, y='word_count', data=df, alpha=0.4, s=15, color='purple')
plt.title('D) Dispersión: Evolución de la Longitud de Diálogos')
plt.xlabel('Secuencia de la Línea (Avance de la Serie)')
plt.ylabel('Cantidad de Palabras')
plt.show()
```

📘 <span style="color:blue">Análisis de Resultados: Scripts de Rick and Morty</span>
📊 <span style="color:blue">A) Histogramas: Forma de la Distribución</span>
🐍 <span style="color:blue">¿La distribución es normal, sesgada o uniforme?</span>
Respuesta: La distribución es sesgada a la derecha (asimetría positiva).

Análisis: No es una curva de campana (normal) ni plana (uniforme). El gráfico muestra una barra muy alta al inicio (izquierda) y una "cola" larga que desciende hacia la derecha.

Razón: Esto indica que la inmensa mayoría de los diálogos son frases cortas (pocas palabras), mientras que las frases muy largas son eventos raros.

📊 <span style="color:blue">B) Boxplots: Outliers y Dispersión</span>
🐍 <span style="color:blue">¿Hay valores atípicos?</span>
Respuesta: Sí, existen múltiples valores atípicos (outliers).

Análisis: Se observan muchos puntos negros fuera de los "bigotes" del diagrama, extendiéndose hacia el lado derecho.

Razón: Estos puntos representan diálogos excepcionalmente largos, como los monólogos científicos de Rick, que se salen del promedio habitual de la serie.

🐍 <span style="color:blue">¿Cómo se ubica la mediana?</span>
Respuesta: La mediana se ubica en un valor bajo (hacia la izquierda).

Análisis: La línea divisoria dentro de la caja está muy pegada al eje Y.

Razón: Esto confirma que, estadísticamente, el "centro" de los datos es una frase corta (entre 5 y 10 palabras), a pesar de que existan discursos largos.

📊 <span style="color:blue">C) Gráfico de Barras: Categorías</span>
🐍 <span style="color:blue">¿Qué categoría tiene más observaciones y por qué?</span>
Respuesta: La categoría predominante es Rick Sanchez, seguida de Morty Smith.

Análisis: La barra correspondiente a Rick es visiblemente más alta que las de los demás personajes.

Razón: Al ser los protagonistas titulares, la trama gira en torno a sus aventuras. Por lógica narrativa, tienen más tiempo en pantalla y más líneas de guion que los personajes secundarios.

📊 <span style="color:blue">D) Gráfico de Dispersión: Tendencias</span>
🐍 <span style="color:blue">¿Se observa alguna tendencia o patrón general?</span>
Respuesta: No se observa una tendencia de crecimiento o decrecimiento clara.

Análisis: La nube de puntos mantiene una densidad constante en la parte inferior a lo largo del eje X (tiempo/índice), sin subir ni bajar diagonalmente.

Razón: El patrón sugiere consistencia. La estructura del guion (muchas frases rápidas intercaladas con algunas largas) se mantiene estable desde el primer episodio hasta el último; el estilo de escritura no cambia drásticamente con el tiempo.
